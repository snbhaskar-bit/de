### Read the custs  comma seprated file from DBFS without header 

### write into DBFS staging/custout with column header , pipe delimited 

```python
schema_str="custid int ,fname string,lname string,age int , prof string"
cust_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/custs",schema=schema_str)

cust_df.show(5)

# write 

# df.write

cust_df.write.mode("overwrite").csv("/Volumes/izwe48catalog/we48db/staging/custout",header=True,sep="|") 
```

## write json format out put 
## javascript object notation 
### semi structure 
### k v  similar to dict

### key are field name / column name  , values -> actual value 

eid | ename | age

100 | raja | 23

### in json 

{"eid":100,"ename":"raja","age":23}





```python
schema_str="custid int ,fname string,lname string,age int , prof string"
cust_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/custs",schema=schema_str)

cust_df.show(5)

# write 

# df.write

cust_df.write.mode("overwrite").json("/Volumes/izwe48catalog/we48db/staging/custjson",compression="gzip") 
```

## xml  - semi structure ,xtensible  markeup language 

### json heavily used , xml now we are not using heavily 

###  markeup -- tag based 


```python
# write xml data 

schema_str="custid int ,fname string,lname string,age int , prof string"
cust_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/custs",schema=schema_str)

cust_df.show(5)

# write 

# df.write

cust_df.write.mode("overwrite").xml("/Volumes/izwe48catalog/we48db/staging/custxml",rowTag="cust") 
```
