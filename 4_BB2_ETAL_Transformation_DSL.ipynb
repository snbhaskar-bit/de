```python
from pyspark.sql.functions import lit ,col,input_file_name,concat,current_date,split

cust_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/BB2/usecase_file_csv",header=True,inferSchema=True)


cust_df.show()
cust_df.count()

# how to take the file name 
# function available to take the file name - input_file_name() -> return a column with filepath 

# cust_df.select("*",input_file_name()).show(5)
cust_df.select("*",col("_metadata.file_path")).show(5,False)

cust_df.select("*",col("_metadata.file_name")).show(15,False)
cust_df.select("*",col("_metadata")).printSchema()




# withColumn  + drop 
cust_enriched_withColumn_df1=cust_df.withColumn("full_name",concat(col("fname"),lit(" "),col("lname"))).withColumn("load_dt",current_date()).withColumn("file_name",split(col("_metadata.file_name"),"\\.")[0]).withColumn("created_by",lit("izuser")).drop("fname","lname")

cust_enriched_withColumn_df2=cust_enriched_withColumn_df1.withColumn("source",split(col("file_name"),"_")[1]).withColumn("data_dt",split(col("file_name"),"_")[2]).drop("file_name")

cust_enriched_withColumn_df3=cust_enriched_withColumn_df2.select("custid","full_name","age","profession","source","load_dt","data_dt","created_by")

cust_enriched_withColumn_df3.write.mode("overwrite").saveAsTable("izwe48catalog.we48db.tbl_cust_detail")
```

```python
cust_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/custs",header=False,sep=",",inferSchema=True).toDF("custid","fname","lname","age","prof")

cust_df.show(2)
```

```python
# write a python function get the age category based on age
# age < 18 -> minor
# age >=18 and age < 45 -> major
# age >=45 and age < 60 -> mid age
# age >=60 -> senior

from pyspark.sql.types import IntegerType,StringType
from pyspark.sql.functions import udf,when
def get_age_category(age):
    if age < 18:
        return "minor"
    elif age >=18 and age < 45:
        return "major"
    elif age >=45 and age < 60:
        return "mid age"
    else:
        return "senior"
    
print(get_age_category(11))
print(get_age_category(45))
print(get_age_category(65))
    
udf_age_cat=udf(get_age_category,StringType())

cust_df2=cust_df.withColumn("age_cat",udf_age_cat(col("age")))

cust_df2.show(15)

# case when using sql expression 


cust_df3=cust_df.selectExpr("*","case when age < 18 then 'minor' when age >=18 and age < 45 then 'major' when age >=45 and age < 60 then 'mid age' else 'senior' end as age_cat")

cust_df3.show(15)

# case when using DSL 

# when(condition,value).otherwise(value) => case when condition then value else value end 
# when(condition,value).when(condition2,value2).otherwise(value)

cust_df3=cust_df.select("*",when(col("age") <18 ,"minor").when((col("age") >=18)&(col("age") <45),"major").when(col("age") <60,"mid age").otherwise("senior").alias("age_cat"))

cust_df3.show()


# when((col("age") >18)&(col("age") <45),"major")
```

```python
cust_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/custs",header=False,sep=",",inferSchema=True).toDF("custid","fname","lname","age","prof")

cust_df.show(5)
```

```python
# total number of records in cust

# df.count - action return - int
# df.groupBy().count()- transformation -> retrun DF

print(cust_df.count())  # returning int result

# total no of records on each prof 
# prof wise count
# select prof,count(1) from tbl group by prof
cust_prof_count_df=cust_df.groupBy("prof").count()

cust_prof_count_df.show()

cust_prof_count_df.count()  # distinct prof count 

# Actor -196
cust_df.filter("prof='Actor'").count()

# age wise count

cust_age_count_df=cust_df.groupBy("age").count()

cust_age_count_df.show()

# prof , age wise count

cust_prof_age_count_df=cust_df.groupBy("prof","age").count()  # datafrme - transformation

cust_prof_age_count_df.show()

# cust level min and maximum age 
from pyspark.sql.functions import min,max,count,avg

cust_df.select(min("age"),max("age")).show()

# prof wise minimum age 

cust_prof_age_min_df=cust_df.groupBy("prof").min("age")

cust_prof_age_min_df.show(50)

# prof wise , min age , max age ,tot_rec

# select prof,min(age),max(age),count(1) from tbl group by prof 

cust_prof_agg_df=cust_df.na.fill("unknown",subset=["prof"]).groupBy("prof").agg(count("*").alias("tot_rec"),min("age").alias("min_age"),max("age").alias("max_age"),avg("age").alias("avg_age"))
cust_prof_agg_df.show(20,False)


# order by 
cust_prof_agg_df.orderBy("tot_rec").show()

cust_prof_agg_df.orderBy("tot_rec",ascending=False).show()


# having clause - filter on group by / finlter on the aggregated res

# prof count ,min,max  prof  which has more than 200

cust_prof_agg_df.filter("tot_rec>200").show()

# select prof,min(age),max(age),count(1) from tbl group by prof having count(1)>200


```

```python

# join syntax 

# df1.join(df2,on=condition,how="left|inner|rght...")
# default join (how) -> inner

# left join

## all records from left table with match from right table
## if no match then null in right table columns

customer_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/joindata/c_customer.csv",sep="|",inferSchema=True,header=True)

trans_df=spark.read.csv("/Volumes/izwe48catalog/we48db/staging/joindata/c_txns.csv",sep="|",inferSchema=True,header=True)

customer_df.show()

trans_df.show()


# left join 
# need all transaction along with matching customer details 

# sql -> join using cid 

# case - 1 join column name is same 
trns_cust_join_df=trans_df.join(customer_df,"cid","left")

trns_cust_join_df=trans_df.join(customer_df,on="cid",how="left")

trns_cust_join_df.show()


# case -2  - join using condition

# columns -> col('col')
# lit('')
# df.colname- col
# select * from tbl a join tbl b on a.cid=b.cid
trns_cust_join_df=trans_df.join(customer_df,trans_df.cid==customer_df.cid,how="left")

trns_cust_join_df.show()

# case 3 - join with diff column name

cust_2_df=customer_df.withColumnRenamed("cid","customer_id")

cust_2_df.show()

trans_df.show()


cust_trans_df=trans_df.join(cust_2_df,trans_df.cid==cust_2_df.customer_id,"left")
cust_trans_df.show()

# case 4 - df alias

# we are setting the table alias for the df , wriring join condition using sql syntax
# select all col from trans and cname from customer

# select * from tbl1 a join tbl2 b on a.id=b.id

from pyspark.sql.functions import expr,col
cust_trans_df=trans_df.alias("a").join(cust_2_df.alias("b"),expr("a.cid=b.customer_id"),"left").select("a.*",col("b.cname"))
cust_trans_df.show()
```

```python
## right join
## all records from rigth table with match from left table
## if no match then null in left table columns


trns_cust_join_df=trans_df.join(customer_df,on="cid",how="right")

trns_cust_join_df.show()
```

```python
# inner join
# only matching records from both tables 

trns_cust_join_df=trans_df.join(customer_df,on="cid",how="inner")

trns_cust_join_df.show()


trns_cust_join_df=trans_df.join(customer_df,on="cid")

trns_cust_join_df.show()
```

```python
# window - group 
# window two things we can mention  - combination of partition and order by 
# partition by - group by - grouping  column
# order by - within the window how you want to order 
```

```python
data =[ ('Lisa', 'Sales', 10000, 35),
          ('Evan', 'Sales', 32000, 38),
          ('Fred', 'Engineering', 21000, 28),
          ('Alex', 'Sales', 30000, 33),
          ('Tom', 'Engineering', 23000, 33),
          ('Jane', 'Marketing', 29000, 28),
          ('Jeff', 'Marketing', 35000, 38),
          ('Paul', 'Engineering', 29000, 23),
          ('Chloe', 'Engineering', 23000, 25)]

df = spark.createDataFrame(data, ['name', 'dept', 'salary', 'age'])
#df.show()

display(df)
```

```python
# ranking function 

# syntax 

# row_number().over(Window.partitionBy("column").orderBy("column"))
# rank().over(Window.partitionBy("column").orderBy("column"))
# dense_rank().over(Window.partitionBy("column").orderBy("column"))

from pyspark.sql.functions import *

from pyspark.sql.window import Window

w_spec=Window.partitionBy("dept").orderBy("salary")



df.withColumn("r_num",row_number().over(w_spec)).withColumn("rank",rank().over(w_spec)).withColumn("d_rank",dense_rank().over(w_spec)).show()






```

```python
# ordering with sal asc and age asc
df.withColumn("rank",rank().over(Window.partitionBy("dept").orderBy("salary","age"))).withColumn("d_rnk",dense_rank().over(Window.partitionBy("dept").orderBy("salary","age"))).show()
```

```python
# Global Rank 

# row_number() -> seq number across table we can use 

df.withColumn("rw_num",row_number().over(Window.orderBy(desc("salary")))).show()
```

```python
# Top N analyses 

# each department wise second highest salary second maximum sal on each department 
# dense_rank 


df2=df.withColumn("rnk",dense_rank().over(Window.partitionBy("dept").orderBy(desc("salary"))))

df2.filter("rnk=2").show()

```
